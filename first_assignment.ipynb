(6-) How do conditional statements work in Python? Illustrate with examples.
--------------------------------------------------------------------------------------------------------------

Ans-
Conditional statements in Python allow you to execute different blocks of code based on certain conditions. 
The main types of conditional statements in Python are if, elif, and else.

Examples:


 * Simple if Statement:
 
 
 age = 18

if age >= 18:
    
    print("You are an adult.")
    
In this example, the condition age >= 18 is checked. Since age is 18, the condition is True, and "You are an adult." is printed.

* if-else Statement:


age = 16

if age >= 18:
    
    print("You are an adult.")
    
else:
    
    print("You are a minor.")

Here, if the condition age >= 18 is True, it prints "You are an adult." Otherwise, it prints "You are a minor." Since age is 16, the condition is False, so "You are a minor." is printed.


* if-elif-else Statement:


score = 85

if score >= 90:
    
    print("Grade: A")
    
elif score >= 80:
    
    print("Grade: B")
    
elif score >= 70:
    
    print("Grade: C")
    
elif score >= 60:
    
    print("Grade: D")
    
else:
    
    print("Grade: F")

In this example, multiple conditions are checked in sequence. The first condition score >= 90 is False, so the next condition score >= 80 is checked, which is True. Therefore, "Grade: B" is printed.

(7) Describe the different types of loops in Python and their use cases with examples.
------------------------------------------

Ans-In Python, loops are used to execute a block of code repeatedly based on a condition. Python supports two main types of loops:

* for Loop
* while Loop

* for Loop:
  
   The for loop is used to iterate over a sequence (like a list, tuple, string, or range) and execute a block of code for each item in the sequence.
   
 Use Cases:
 
 * Iterating through a collection of items (e.g., list, tuple).
 * Generating a sequence of numbers using the range() function.
*  Processing elements in a string or list.

Examples:

* Iterating Through a List:

fruits = ["apple", "banana", "cherry"]

for fruit in fruits:

    print(fruit)
    
* This example prints each fruit in the list.


* while Loop

The while loop repeatedly executes a block of code as long as a specified condition is True.

Use Cases:

* Repeating a task until a condition changes.
* Waiting for a specific condition to be met before proceeding.
* Implementing loops with complex termination conditions.

Examples:

* Simple while Loop:

count = 0

while count < 5:
    print(count)
    count += 1

* The loop prints numbers from 0 to 4. It continues executing until count is no longer less than 5.
